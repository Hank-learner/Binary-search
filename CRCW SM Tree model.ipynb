# CRCW SM Tree model Binary Search
1. N leaf processors (can also be implemented using threads)
2. n elements in sorted Array

```python
import multiprocessing
from joblib import Parallel, delayed
import time
import random
```

<b>getting count of processors, real and virtual</b>

```python
num_processors = multiprocessing.cpu_count()
print( num_processors )
```

<b>getting a randomised sorted array</b>

```python
sortedArray=[]
n=100000

sortedArray.append(1)
for i in range(1, n): 
    ele = random.randint(sortedArray[i-1],sortedArray[i-1]+10)
    sortedArray.append(ele)
    
x = sortedArray[random.randint(1,n-1)]

print(sortedArray)
```

<b>binary search algorithm</b>
1. input:

    1. array
    
    2. left index
    
    3. right index
    
    4. search value
    
2. output:

    1. index

```python
found = False
index = -1

def binarySearch (i):
    global x, index, found
    l = int(i * (n / 4)); 
    r = int((i + 1) * (n / 4));
    
    while l <= r and not found: 
        mid = int(l + (r - l) // 2) 
        
        if sortedArray[mid] == x: 
            index = mid
            found = True
            print("Found by processor", i)
            break
        elif sortedArray[mid] < x: 
            l = mid + 1
        else: 
            r = mid - 1
```

Processors are used in tree, here leaf nodes find out the index value of the element x , hence there is no use for parent processors , As compared to matrix multiplication, leaf nodes perform multiplication, the next level parent from bottom computes addition for the data and hence root node is the output final matrix, but here in binary search, the leaf processors get the result in a single stage, thus there is no use for parent processor, thus this method is equal to simd vector processors

selecting 4 leaf processors and executing in parallel

```python
use_processors = 4 # num_processors

start = time.time()
Parallel(n_jobs = 4,prefer="threads")(delayed(binarySearch)(i) for i in range(0, use_processors))
end = time.time()

if index != -1: 
    print ("Element %d is present at index % d" % (x,index)) 
else: 
    print ("Element %d is not present in array" % (x))

print(f"Runtime of the program is {end - start}")
```

Thus the binary search is executed in parallel and time taken is measured
